```python
df = pd.read_csv("http://cowid.netlify.com/data/full_data.csv")
```

```python
plt.plot(df[df['location']=='Worldwide']['date'],df[df['location']=='Worldwide']['new_deaths'])
```

```python
df['rate'] = df.total_cases / df.total_cases.shift(1)
#df['newcase_ratio'] = df.new_cases / df.new_cases.shift(1)
```

```python
country = 'China'
n = 100

fig, ax1 = plt.subplots(figsize=(20,10))

t = df[df['location']==country][df['total_cases'] > n]['date'].iloc[1:]
data1 = df[df['location']==country][df['total_cases'] > n]['new_cases'].iloc[1:]
data2 = df[df['location']==country][df['total_cases'] > n]['rate'].iloc[1:]

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('new cases', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('DoD total cases ratio', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
